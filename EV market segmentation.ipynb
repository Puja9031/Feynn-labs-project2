### Description Of Columns
-Age:The age of the individual.

-Profession: The profession (e.g., Salaried, Business).

-Marital Status: Whether the individual is single or married.

-Education: Level of education (e.g., Post Graduate, Graduate).

-No of Dependents: Number of dependents in the family.

-Personal Loan: Whether the individual has a personal loan.

-House Loan: Whether the individual has a house loan.

-Wife Working: Indicates if the spouse is working.

-Salary: The salary of the individual.

-Wife Salary: The salary of the spouse, if applicable.

-Total Salary: Combined family income.

-Make: The car make or model the individual bought.

-Price: The price of the car purchased.

### Applying K-Means clustering

```python
from sklearn.preprocessing import StandardScaler

# Select features for clustering (example)
features = df[['Age', 'Total_Salary', 'Car_Price','No_of_Dependents']]  # Replace with actual feature names
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
wcss = []  # Within-cluster sums of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_k = 2  # Replace with the optimal number found using the elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['Car_Price'], hue=df['Total_Salary'], palette='Set2', s=100)
plt.title('K-Means Clustering')
plt.xlabel('Age')
plt.ylabel('Car_Price')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['Car_Price'], df['Total_Salary'], df['Age'], 
           c=df['Cluster'], cmap='viridis', s=100)

# Labels and title
ax.set_title('3D K-Means Clustering')
ax.set_xlabel('Car_Price')
ax.set_ylabel('Total_Salary')
ax.set_zlabel('Age')

# Show plot
plt.show()
```
